The data about ЗАЕС after 2022 is missing, that explains the 13 missing values in each column.

Also co_60_dump at ХАЕС is always missing.

So let's delete rows about ЗАЕС after 2022.

```python
# Create masks for each condition
mask1 = (df['station'] == 'ЗАЕС') & (df['year'] > 2022)
mask2 = (df['station'] == 'ЗАЕС') & (df['year'] == 2022) & (df['quarter'] == 4)

# Combine masks with OR (|) and then invert with NOT (~)
filtered_df = df[~(mask1 | mask2)]
filtered_df
```

```python
df.dtypes
```

There is some columns with '<0.01' or something like that, it prevents from treating these columns as numerical

So lets delete '<' and convert types

```python
# First remove '<' symbols
df = df.replace('<', '', regex=True)

# Then convert all columns to numeric where possible
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.head()
```

Data is cleaned, save it for further use
